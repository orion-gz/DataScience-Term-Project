### Feature Engineering

#### Feature Creation

##### Tags to Binary Data

```python
# Make Tag list and number of tag from preprocessed data frame
tags = df_model['tags'].dropna()
tags_list = []

for tag in tags:
  items = tag.split(', ')
  for item in items:
    tags_list.append(item.strip())

# Counter for preprocessed data's Top 200 tags
p_tags_list = Counter(tags_list).most_common(200)
# get tags name list
p_tags_name_list = [tag for tag, count in p_tags_list]
# get number of each tag
p_tags_num_list = [count for tag, count in p_tags_list]
```

```python
print(p_tags_name_list)
print(p_tags_num_list)
```

```python
# tags to binary data
tags_columns_data = {}
for tag in p_tags_name_list:
  col = f"tag_{tag.replace(' ', '_').lower()}"
  col_values = []
  for tag_list in df_model['tags']:
    if tag in tag_list:
      col_values.append(1)
    else:
      col_values.append(0)
    tags_columns_data[col] = col_values

df_model_tags = pd.DataFrame(tags_columns_data, index=df_model.index)
df_model = pd.concat([df_model, df_model_tags], axis=1)
```

##### Genres to Binary Data

```python
# Make genre list and number of genre from preprocessed data frame
# variable name p means preprocessed data
p_genres = df_model['genres'].dropna()
p_genres_list = []

for p_genre in p_genres:
  items = p_genre.split(', ')
  for item in items:
    p_genres_list.append(item.strip())

# Counter for preprocessed data's genres
p_genres_list = Counter(p_genres_list)
# get genres name list
p_genres_name_list = [genre for genre in p_genres_list]
# get number of each genre
p_genres_num_list = [count for genre, count in p_genres_list.items()]
```

```python
print(p_genres_name_list)
print(p_genres_num_list)
```

```python
# genres(Categorical data) into binary data(manually)
genres_columns_data = {}
for genre in p_genres_name_list:
  col = f"genre_{genre.replace(' ', '_').lower()}"
  col_values = []
  for genre_list in df_model['genres']:
    if genre in genre_list:
      col_values.append(1)
    else:
      col_values.append(0)
  genres_columns_data[col] = col_values

df_model_genres = pd.DataFrame(genres_columns_data, index=df_model.index)
df_model = pd.concat([df_model, df_model_genres], axis=1)
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd

# SelectKBest
feature_selected = featureScores.nlargest(11, 'score')['features'].copy().tolist()
# If important features are not in selected feature add manually
# because almost features are binary.
# Ensure 'metacritic' and 'playtime' are in the selected features if not already
if 'metacritic' not in feature_selected:
    feature_selected.append('metacritic')
if 'playtime' not in feature_selected:
    feature_selected.append('playtime')

final_df_features = df_model[feature_selected].copy()
final_df_target = df_model['rating']

X_train, X_test, y_train, y_test = train_test_split(final_df_features, final_df_target,
                                                    test_size=0.2,
                                                    random_state=42,
                                                    shuffle=True
                                                    )

X_train['meta * atmospheric'] = X_train['metacritic'] * X_train['tag_atmospheric']
X_test['meta * atmospheric'] = X_test['metacritic'] * X_test['tag_atmospheric']

X_train['meta * great_soundtrack'] = X_train['metacritic'] * X_train['tag_great_soundtrack']
X_test['meta * great_soundtrack'] = X_test['metacritic'] * X_test['tag_great_soundtrack']

X_train['meta * soundtrack'] = X_train['metacritic'] * X_train['tag_soundtrack']
X_test['meta * soundtrack'] = X_test['metacritic'] * X_test['tag_soundtrack']

X_train['meta * story_rich'] = X_train['metacritic'] * X_train['tag_story_rich']
X_test['meta * story_rich'] = X_test['metacritic'] * X_test['tag_story_rich']

X_train['meta * story'] = X_train['metacritic'] * X_train['tag_story']
X_test['meta * story'] = X_test['metacritic'] * X_test['tag_story']

X_train['meta * classic'] = X_train['metacritic'] * X_train['tag_classic']
X_test['meta * classic'] = X_test['metacritic'] * X_test['tag_classic']

X_train['meta * singleplayer'] = X_train['metacritic'] * X_train['tag_singleplayer']
X_test['meta * singleplayer'] = X_test['metacritic'] * X_test['tag_singleplayer']

X_train['meta * openworld'] = X_train['metacritic'] * X_train['tag_open_world']
X_test['meta * openworld'] = X_test['metacritic'] * X_test['tag_open_world']

X_train['meta * third_person'] = X_train['metacritic'] * X_train['tag_third_person']
X_test['meta * third_person'] = X_test['metacritic'] * X_test['tag_third_person']

X_train['meta * horror'] = X_train['metacritic'] * X_train['tag_horror']
X_test['meta * horror'] = X_test['metacritic'] * X_test['tag_horror']

features_to_poly = [col for col in X_train.columns if 'meta *' in col or col in ['metacritic', 'playtime']]

poly_feature = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)

X_train_poly_transformed = poly_feature.fit_transform(X_train[['metacritic', 'playtime']])
X_test_poly_transformed = poly_feature.transform(X_test[['metacritic', 'playtime']])

poly_feature_names = poly_feature.get_feature_names_out()

X_train_poly_transformed = pd.DataFrame(X_train_poly_transformed, columns=poly_feature_names, index=X_train.index)
X_test_poly_transformed = pd.DataFrame(X_test_poly_transformed, columns=poly_feature_names, index=X_test.index)

poly_model = LinearRegression()
poly_model.fit(X_train_poly_transformed, y_train)
y_pred = poly_model.predict(X_test_poly_transformed)

score = poly_model.score(X_test_poly_transformed, y_test)
print(score)
```
